```python
f_class_Age=a.groupby('Pclass')['Age'].mean()
f_class_Age=pd.DataFrame(f_class_Age)
f_class_Age.plot.bar(y='Age')
plt.title("average age of passengers by class",fontsize=12)
plt.ylabel("Age (years)",fontsize=12)
plt.xlabel("passenger class",fontsize=12)
```

```python
import numpy as np
import pandas as pd
b=list(f_class_Age['Age'])
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
     if Pclass==1:
        return b[0]
     elif Pclass==2:
        return b[1]
     else:
        return b[2]
  else:
    return Age
```

```python
a['Age']=a[['Age','Pclass']].apply(impute_age,axis=1)
d=a.describe()
dt=d.T
dt.plot.bar(y='count')
plt.title("bar plot of the count",fontsize=6)
```
